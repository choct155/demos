# PyMC3 Demo: Intro with Simulated Data

In this notebook, we are just trying to get a feel for the syntax and user experience. We'll simulate some straightforward model that has known weights and use `pymc3` to execute Ordinary Least Squares regression.

```python
import numpy as np
import xarray as xr
import pymc3 as pm
import seaborn as sb
from dataclasses import dataclass
import pandas as pd

%pylab inline
```

## Data Simulation

As previously stated, we'll just simulate some data with known weights.

```python
@dataclass
class SimOLSData:
    beta: np.ndarray
    sigma: float
    n: int = 1000
    
    def __post_init__(self) -> None:
        num_weights: int = self.beta.shape[0]
        self.x: np.ndarray = np.random.normal(size=num_weights * self.n).reshape(self.n, num_weights)
        self.y_hat: np.ndarray = np.squeeze(np.matmul(self.x, self.beta[..., np.newaxis]))
        self.y: np.ndarray = self.y_hat + np.random.normal(scale=self.sigma, size=self.n)
        y: xr.DataArray = xr.DataArray(
            data=self.y,
            coords=dict(obs=np.arange(self.n)),
            dims=["obs"]
        )
        x: xr.DataArray = xr.DataArray(
            data=self.x,
            coords=dict(obs=np.arange(self.n), reg=[f"x{i}" for i in range(num_weights)]),
            dims=["obs", "reg"]
        )
        self.ds: xr.Dataset = xr.Dataset({"y": y, "x": x})
        
    def data_as_df(self) -> pd.DataFrame:
        df: pd.DataFrame = self.ds["x"].to_dataframe().unstack("reg")
        df.columns = [col[1] for col in df.columns]
        df["y"] = self.ds["y"].data
        sorted_columns: List[str] = ["y"] + [col for col in df.columns if col != "y"]
        return df[sorted_columns]
    
    def plot_data(self) -> pd.DataFrame:
        df: pd.DataFrame = self.data_as_df()
        pairplot: sb.PairGrid = sb.pairplot(data=df)
        return pairplot
        
sols: SimOLSData = SimOLSData(beta=np.array([1., -2., 3.]), sigma=2.)
sols.ds
```

```python
sols.plot_data()
```

## Model Estimation

Now we get to the heart of this. Can we recover our weights? It appears that `pymc3` builds its models by way of a `ContextManager`.

```python
m1: pm.Model = pm.Model()

with m1:
    beta: pm.MvNormal = pm.MvNormal(
        "weights", 
        mu=np.eye(sols.beta.shape[0]),
        cov=np.eye(sols.beta.shape[0])
    )
    sigma: pm.Uniform = pm.Uniform(
        lower=1.,
        upper=10.
    )
    
    mu: np.array = pm.math.det(sols.x, beta)
    print(mu)
    
```
